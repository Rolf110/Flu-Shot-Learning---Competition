```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('./Data/training_set_features.csv')
df2 = pd.read_csv('./Data/training_set_labels.csv')
```

```python
df2.head()
```

```python
df = pd.merge(df1, df2, on='respondent_id')
```

```python
df
```

```python
df.info()
```

```python
#number of NaNs in each column
df.isna().sum()
```

```python
# drop some features due to large number of NaNs in them
#df.drop(['health_insurance', 'employment_industry', 'employment_occupation'], axis=1, inplace=True)
```

```python
#fill NaNs in columns with the largest number of NaNs
unknown_filled_features = ['health_insurance', 
                           'employment_industry', 
                           'employment_occupation', 
                           'income_poverty',
                           'education']

df[unknown_filled_features] = df[unknown_filled_features].fillna('unknown')
```

```python
#drop rows with NaNs
dropna_features = ['h1n1_concern',
                   'h1n1_knowledge',
                   'doctor_recc_h1n1', 
                   'doctor_recc_seasonal',
                   'chronic_med_condition',
                   'health_worker']

df = df.loc[df[dropna_features].dropna().index]
```

```python
#NaNs in binary columns
bin_col = [col for col in df if 
               df[col].value_counts().index.isin([0,1]).all()]
df[bin_col].isna().sum()
```

```python
#fill NaNs in categorical columns with mode 
cat_col = df.select_dtypes('object')
for col in cat_col:
    df[col] = df[col].fillna(value = df[col].mode().max())
```

```python
df[list(cat_col)].isna().sum()
```

```python
#fil NaNs in num columns with median
num_col = df.select_dtypes('number')
for col in num_col:
    df[col] = df[col].fillna(value = df[col].median())
```

```python
df.isna().sum()
```

Feature employment_industry and employment_occupation include too much NaNs - so it's better to drop them

Plot vaccination rates (for seasonal and h1n1 vaccines) for each categorical feature

```python
cols_lineplot = ['age_group',
 'education',
 'race',
 'sex',
 'income_poverty',
 'marital_status',
 'rent_or_own','census_msa', 'hhs_geo_region']
```

```python
df.select_dtypes(include='object')
```

```python
def plot_rates_h(x,vac):
    a = df.groupby(x)[vac].sum()
    b = df[x].value_counts().sort_index()
    if list(a.index) == list(b.index):
        rate = list()
        for i in range(len(a)):
            rate.append(a.values[i]/b.values[i])
    pl = list()
    pl.append(list(a.index))
    pl.append(rate)
    return pl
```

```python
fig, ax = plt.subplots(int(len(cols_lineplot)), 2, figsize=(20, len(cols_lineplot)*4))
for idx, col in enumerate(cols_lineplot):
    sns.lineplot(x=plot_rates_h(col, 'h1n1_vaccine')[0], y=plot_rates_h(col, 'h1n1_vaccine')[1], ax=ax[idx, 0], marker='o').set_xlabel(col)
    sns.lineplot(x=plot_rates_h(col, 'seasonal_vaccine')[0], y=plot_rates_h(col, 'seasonal_vaccine')[1], ax=ax[idx, 1], marker='o').set_xlabel(col)
ax[0, 0].legend(
    loc='lower center', bbox_to_anchor=(0.5, 1), title='h1n1_vaccine'
)
ax[0, 1].legend(
    loc='lower center', bbox_to_anchor=(0.5, 1), title='seasonal_vaccine'
) 
plt.tight_layout()
```
